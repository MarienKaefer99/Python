# 5. Regression

 = method to predict values of one numerical variable from values of another numerical variable

## 5.1 Linear regression 

* straight line to predict response variable
* $Y = a + b\cdot X$
    * a = intercept
    * b = slope = 
    $\dfrac{\sum\nolimits_{i}(X_i-\overline{X}) \cdot (Y_i - \overline{Y})}{\sqrt{\sum\nolimits_{i}(X_i-\overline{X})^2}}$
    
**Assumptions:**
* linear relationship between variables
* For each value of X, the distribution of possible Y-values is normal
* Variance of Y-values is the same at all values of X
* At each value of X, the Y measurements represent a random sampple from the population of possible Y values
* No assumptions about distribution of X values

**Effect of measurement errors:**
* error in Y $\rightarrow$ Moving the data along the Y-axis $\rightarrow$ no effect on regression
* error in X $\rightarrow$ Pushing the data together or pulling them apart $\rightarrow$ effect on regression (slope)

### 5.1.1 Residuals

=  Difference between the actual observed values $Y_i$ and the predicted values $\hat{Y}_i$ of a statistical analysis
* $ \Delta r_i = Y_i - \hat{Y}_i $
* $\hat{Y}_i = a + b \cdot X_i$

**Residual plot**
* Residuals can be plot in Scatter plot
* shows the size of the deviation
* linear relationship, if:
    * symmetric cloud of points above and below horizontal line at zero
    * higher density of points close to the line than away 
    * no curvature
    * Approximately equal variance of points above and below the line at all values of X

**Interpretation of residual plot**
* Homoscedaticity = same variance for each dot
    * data is linear
* Heteroscedasticity = varianxe varies for each dot
    * data is not linear
    * solution: transform data (log or square root)

# 7. Sampling distribution

= probability distribution of all values for an estimate that we might obtain when we

**Standard error of the mean:**
* $\sigma_{\overline{Y}} = \dfrac{s}{\sqrt{n}}$
    * $s$ = standard deviation of the sample
* In general, the standard error of a sample parameters describes the standard deviation of the sampling distribution.
    
**Confidence interval:**
* gibt einen Vertrauensbereich an, in dem der wahre Mittelwert mit einer x-%igen Wahrscheinlichkeit liegt.
* Ich bin zu X % sicher, dass der wahre Mittelwert in diesem Intervall liegt
    
<u>Student's t-distribution
* used, when:
    * unknown variance
    * smaller sample size
* $Z = \dfrac{\overline{Y} - \mu}{\sigma_{\overline{Y}}}$
    * $\overline{Y}$ = sample mean
    * $\mu$ = population mean
* degrees of freedom: $n-1$
    
<u> $x^2$-distribution
* for larger sample size it converts to a normal distribution

* Confidence interval of the slope:
